### Something is making the average AQI look worse than it is

A data quality report flags that a few AQI readings in the dataset are implausibly high — values that no monitoring station should realistically record. If left in, these values will distort every statistic and model built on this data. The team lead asks you to "handle the extreme values properly" but does not tell you how.
Identify whether extreme values exist, quantify them, decide on an appropriate treatment method, apply it, and demonstrate that the data is now cleaner. Justify every decision you make.

Step 1: Detect and Quantify (The IQR Method)

The most standard and statistically robust way to identify outliers is the Interquartile Range (IQR) method. It calculates the "normal" spread of your data and sets a mathematical upper bound. Anything beyond that bound is an outlier.

```python
import numpy as np

# 1. Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = city_clean['AQI'].quantile(0.25)
Q3 = city_clean['AQI'].quantile(0.75)

IQR = Q3 - Q1

# 3. Define the Upper Bound (Formula: Q3 + 1.5 * IQR) { > 0
upper_bound = Q3 + 1.5 * IQR

# 4. Quantify how many values are above this bound
outliers = city_clean[city_clean['AQI'] > upper_bound]
print(f"The mathematical upper bound is: {upper_bound:.2f}")
print(f"Number of implausibly high AQI records found: {len(outliers)}")
```

Step 2: Treat the Outliers (Capping / Winsorization)
Treatment Choice:

We should Cap the outliers rather than drop them. Why? If we drop them, we lose an entire day's worth of data for that city just because the AQI sensor malfunctioned. By capping (replacing any value above the upper bound with the upper bound), we preserve the fact that it was a very polluted day, but we stop the impossible 2000+ values from destroying our average.

```python
# Create a new column or overwrite the existing one to cap the values
# np.where works like an IF statement: IF value > bound, REPLACE with bound, ELSE keep original
city_clean['AQI_Cleaned'] = np.where(city_clean['AQI'] > upper_bound, upper_bound, city_clean['AQI'])

print("Treatment applied: Capping at upper bound.")
```

Step 3: Visual Comparison

Now we need to prove to the team lead that our treatment worked using a side-by-side boxplot.

```python
plt.figure(figsize=(12, 5))

# Plot 1: Before Treatment
plt.subplot(1, 2, 1)
sns.boxplot(x=city_clean['AQI'], color='lightcoral')
plt.title('Before Treatment (Original)')

# Plot 2: After Treatment
plt.subplot(1, 2, 2)
sns.boxplot(x=city_clean['AQI_Cleaned'], color='lightgreen')
plt.title('After Treatment (Capped)')

plt.tight_layout()
plt.show()
```

#### Can the two datasets talk to each other?

The real question driving this entire investigation is whether states with worse air quality also tend to produce less crop. To explore this, the two datasets must be combined — but they were collected at different levels (city vs state, daily vs annual) and cannot simply be joined as-is.

Transformation reasoning:
1. city_clean has city-level, daily rows.
2. crop_clean has state-level, annual rows.
3. To merge, both must share the same granularity: state.
4. For city data: aggregate daily AQI → grouped by State 
5. For crop data: aggregate production → total or average per State.
6. The merge key will be ['State']

```python
# aggregation city data to state-year level

city_by_state = city_clean.groupby('State')['AQI'].mean().reset_index()
```

```python
# aggregation crop data to state-year level

crop_by_state = crop_clean.groupby('State_Name')[['Area', 'Production']].mean().reset_index()
crop_by_state.rename(columns={'State_Name': 'State'}, inplace=True)
```

```python
# merging

merged = pd.merge(city_by_state, crop_by_state, on='State', how='inner')

print(merged.shape)
merged.head()
```

```python
plt.ioff()

fig, ax = plt.subplots(
    figsize=(8,6)
)
```

```python
# Creating figure for correlation matrix

corr = merged[['AQI', 'Area', 'Production']].corr()

sns.heatmap(
    corr, 
    annot=True, 
    cmap='coolwarm', 
    center=0,
    ax=ax
)
plt.title('Correlation Matrix: Air Quality vs Crop Metrices')

plt.ion()
plt.tight_layout()
plt.show()
```

>"The city data covers 2015–2020 while the crop data covers 1997–2015, so the time periods barely overlap. To create a meaningful comparison, we aggregated both datasets to the state level (averaging AQI across all available years, and averaging crop production across all available years), then merged on the State column. This sacrifices temporal resolution but gives us enough matched data points to explore the AQI-agriculture relationship."
