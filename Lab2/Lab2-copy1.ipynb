```python
city_clean = city.copy()
city_clean.drop(columns = ['Xylene'], inplace=True)

city_clean.info()
```

Pollutants & AQI: These are time-series readings. We should sort by City and Date and use linear interpolation. This estimates missing values based on the days immediately before and after for that specific city.

```python
# City Day Strategy

city_clean['Date'] = pd.to_datetime(city_clean['Date'])
print(city_clean['Date'].dtype)
city_clean = city_clean.sort_values(by=['City', 'Date'])
print(city_clean[['Date', 'City']])
```

```python
# Grouping by cities and interpolate missing time-series data

pollutant_cols = [
    'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'AQI'
]
for col in pollutant_cols:
    city_clean[col] = city_clean.groupby('City')[col].transform(lambda x: x.interpolate(method='linear', limit_direction='both'))
```

```python
city_clean['AQI'] = city['AQI']
city_clean['AQI'].unique()
```

### Something is making the average AQI look worse than it is

A data quality report flags that a few AQI readings in the dataset are implausibly high — values that no monitoring station should realistically record. If left in, these values will distort every statistic and model built on this data. The team lead asks you to "handle the extreme values properly" but does not tell you how.
Identify whether extreme values exist, quantify them, decide on an appropriate treatment method, apply it, and demonstrate that the data is now cleaner. Justify every decision you make.

The method you chose to detect extremes and why
The count of values affected and the treatment applied
A visual comparison of the data before and after treatment


Step 1: Detect and Quantify (The IQR Method)

The most standard and statistically robust way to identify outliers is the Interquartile Range (IQR) method. It calculates the "normal" spread of your data and sets a mathematical upper bound. Anything beyond that bound is an outlier.

```python
import numpy as np

# 1. Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = city_clean['AQI'].quantile(0.25)
Q3 = city_clean['AQI'].quantile(0.75)

IQR = Q3 - Q1

# 3. Define the Upper Bound (Formula: Q3 + 1.5 * IQR) { > 0
upper_bound = Q3 + 1.5 * IQR

# 4. Quantify how many values are above this bound
outliers = city_clean[city_clean['AQI'] > upper_bound]
print(f"The mathematical upper bound is: {upper_bound:.2f}")
print(f"Number of implausibly high AQI records found: {len(outliers)}")
```

Step 2: Treat the Outliers (Capping / Winsorization)
Treatment Choice:

We should Cap the outliers rather than drop them. Why? If we drop them, we lose an entire day's worth of data for that city just because the AQI sensor malfunctioned. By capping (replacing any value above the upper bound with the upper bound), we preserve the fact that it was a very polluted day, but we stop the impossible 2000+ values from destroying our average.

```python
# Create a new column or overwrite the existing one to cap the values
# np.where works like an IF statement: IF value > bound, REPLACE with bound, ELSE keep original
city_clean['AQI_Cleaned'] = np.where(city_clean['AQI'] > upper_bound, upper_bound, city_clean['AQI'])

print("Treatment applied: Capping at upper bound.")
```

Step 3: Visual Comparison

Now we need to prove to the team lead that our treatment worked using a side-by-side boxplot.

```python
plt.figure(figsize=(12, 5))

# Plot 1: Before Treatment
plt.subplot(1, 2, 1)
sns.boxplot(x=city_clean['AQI'], color='lightcoral')
plt.title('Before Treatment (Original)')

# Plot 2: After Treatment
plt.subplot(1, 2, 2)
sns.boxplot(x=city_clean['AQI_Cleaned'], color='lightgreen')
plt.title('After Treatment (Capped)')

plt.tight_layout()
plt.show()
```

#### Is India's air getting better or worse over time?

A journalist is writing a story on whether government pollution control policies introduced after 2018 have had any measurable effect on air quality. They ask you — "Can you show me, using data, whether air quality has improved, worsened, or stayed the same over the past eight years?"

Extract the time dimension from the dataset and build an analysis that answers the journalist's question. Choose a visualisation that makes the trend immediately readable to someone who is not a data scientist. Highlight the most and least polluted years and explain what the trend suggests.

Your plot with the trend clearly visible and key years highlighted

1. Extract the time dimension (Year) from the Date column


```python
city_clean['Year'] = city_clean['Date'].dt.year
```

2. Calculate the average AQI for each year to find the trend

```python
yearly_aqi = city_clean.groupby('Year')['AQI'].mean().reset_index()
```

3. Identify the most and least polluted years

```python
most_polluted = yearly_aqi.loc[yearly_aqi['AQI'].idxmax()]
least_polluted = yearly_aqi.loc[yearly_aqi['AQI'].idxmin()]
```

4. Create a readable, non-technical visualization and plot the main trend line

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_aqi,
x='Year', y='AQI', marker='o', linewidth=3, 
color='#7F8C8D', label='Average AQI Trend')
```

Highlight the highest (worst) and lowest (best) years

```python
plt.scatter(most_polluted['Year'], most_polluted['AQI'], color='#E74C3C', s=200, zorder=5, 
            label=f"Most Polluted ({int(most_polluted['Year'])})")
plt.scatter(least_polluted['Year'], least_polluted['AQI'], color='#2ECC71', s=200, zorder=5, 
            label=f"Least Polluted ({int(least_polluted['Year'])})")
```

Format the plot for a general audience (Journalist-friendly)

```python
plt.title('Air Quality Trend (2015 - 2020)', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Air Quality Index (AQI)\n← Lower means cleaner air', fontsize=12)
plt.xticks(yearly_aqi['Year']) # Show whole years on the x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(fontsize=11)
plt.tight_layout()
```

Answer to the journalist

Our analysis reveals that 2015 was the most polluted year, with the highest average Air Quality Index (AQI) levels. However, in recent years, there has been a noticeable and measurable downward trend in pollution. By 2020, the data recorded a dramatic drop in the AQI, making it the cleanest and least polluted year on record. While this significant improvement suggests that recent pollution control policies have had a positive effect, the steep drop in 2020 also likely reflects the unique reduction in emissions during the lockdowns, proving that sustained, reduced emissions directly result in better air quality.
