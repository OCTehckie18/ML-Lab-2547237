```python
city_clean = city.copy()
city_clean.drop(columns = ['Xylene'], inplace=True)

city_clean.info()
```

Pollutants & AQI: These are time-series readings. We should sort by City and Date and use linear interpolation. This estimates missing values based on the days immediately before and after for that specific city.

```python
# City Day Strategy

city_clean['Date'] = pd.to_datetime(city_clean['Date'])
print(city_clean['Date'].dtype)
city_clean = city_clean.sort_values(by=['City', 'Date'])
print(city_clean[['Date', 'City']])
```

```python
# Grouping by cities and interpolate missing time-series data

pollutant_cols = [
    'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'AQI'
]
for col in pollutant_cols:
    city_clean[col] = city_clean.groupby('City')[col].transform(lambda x: x.interpolate(method='linear', limit_direction='both'))
```

```python
city_clean['AQI'] = city['AQI']
city_clean['AQI'].unique()
```

A Hybrid Approach
You can combine your idea with time-series interpolation to get the best of both worlds!

First Pass (Time-series): Use linear interpolation grouped by City (preserves seasons and city-specific trends).
Second Pass: Create the State column. For any AQI values that are still null after interpolation, fill them with the mean AQI of their respective State.

```python
# resetting the AQI Column

city_clean['AQI'] = city['AQI']
```

```python
# pass 1: time-series linear interpolation grouped by city
city_clean['AQI'] = city_clean.groupby('City')['AQI'].transform(
    lambda x: x.interpolate(method='linear', limit_direction='both')
)
```

```python
print("Nulls after interpolation:", city_clean['AQI'].isnull().sum())
```

```python
city_clean["AQI"].unique()
```

```python

# Dictionary mapping cities to states
city_to_state = {
    'Ahmedabad': 'Gujarat', 'Delhi': 'Delhi', 'Visakhapatnam': 'Andhra Pradesh',
    'Amaravati': 'Andhra Pradesh', 'Amritsar': 'Punjab', 'Bengaluru': 'Karnataka',
    'Bhopal': 'Madhya Pradesh', 'Brajrajnagar': 'Odisha', 'Chandigarh': 'Chandigarh',
    'Chennai': 'Tamil Nadu', 'Coimbatore': 'Tamil Nadu', 'Ernakulam': 'Kerala',
    'Gurugram': 'Haryana', 'Guwahati': 'Assam', 'Hyderabad': 'Telangana',
    'Jaipur': 'Rajasthan', 'Jorapokhar': 'Jharkhand', 'Kochi': 'Kerala',
    'Kolkata': 'West Bengal', 'Lucknow': 'Uttar Pradesh', 'Mumbai': 'Maharashtra',
    'Patna': 'Bihar', 'Shillong': 'Meghalaya', 'Talcher': 'Odisha',
    'Thiruvananthapuram': 'Kerala', 'Aizawl':'Mizoram'
}

# Create the State column
city_clean['State'] = city_clean['City'].map(city_to_state)

# Check if any cities were missed in the mapping
print("Unmapped cities:", city_clean[city_clean['State'].isnull()]['City'].unique())
```

```python
# Calculate the mean AQI for each state
state_aqi_means = city_clean.groupby('State')['AQI'].transform('mean')

# 2nd Pass: Fill any remaining NaNs with the State's mean AQI
city_clean['AQI'].fillna(state_aqi_means, inplace=True)

# Final check
print("Nulls after State-mean fallback:", city_clean['AQI'].isnull().sum())
```

I chose to impute missing AQI values using the mean AQI of the respective State. While time-series interpolation is an option, it can create artificial repeating values over large gaps. Grouping by State and taking the mean ensures that missing values are filled with a realistic baseline for that specific region without making assumptions about daily trends.
